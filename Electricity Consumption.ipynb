```python
# Check out the data if its fit for use using visuals

#Average solar electricity generation and average electricity usage for each hour in a day, i.e. the average over all amounts for 1 am, and each subsequent hour within a 24-hour period.

#• investigating any significant outliers.

#• making any corrections that are needed.


#Calculating hrly average

hourly_electricity_avg = df.groupby("Hour").agg({
    
    'Solar electricity generation (kWh)': 'mean',
    'Electricity usage (kWh)': 'mean'
}).reset_index()

```

```python
#Electricity usage seems to spike at 12 noon  by a significant margin highlighting potential outliers


#use z_score and check for outliers z_score>3


import numpy as np
from scipy import stats


z_score_solargeneration = np.abs(stats.zscore(df["Solar electricity generation (kWh)"])) #.abs() is used to give the absolute value of the numbers
z_score_electricityusage = np.abs(stats.zscore(df["Electricity usage (kWh)"]))


#Finding outliers z_score> 3

outliers_solar = df[z_score_solargeneration > 3]
outliers_usage = df[z_score_electricityusage > 3]

print("Solar Generation Outliers:")
print(outliers_solar.sort_values('Solar electricity generation (kWh)', ascending=False).head())

print("\
Electricity Usage Outliers:")
print(outliers_usage.sort_values('Electricity usage (kWh)', ascending=False).head())

# Create box plots to visualize the distribution and outliers
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.boxplot(df['Solar electricity generation (kWh)'])
plt.title('Solar Generation Distribution')
plt.ylabel('kWh')

plt.subplot(1, 2, 2)
plt.boxplot(df['Electricity usage (kWh)'])
plt.title('Electricity Usage Distribution')
plt.ylabel('kWh')

plt.tight_layout()
plt.show()
```

```python
#Remove outliers using Interquartile range(IQR)

def remove_outliers(df, column):
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3-Q1
    
    lower_bound = Q1- 1.5* IQR
    upper_bound = Q3+ 1.5* IQR
    
    return df[ (df[column] >= lower_bound) & (df[column]<= upper_bound)]


#clean both columns

df_no_outliers = df.copy()# first copy your dataframe so as to retain the original df just incase

df_no_outliers = remove_outliers(df_no_outliers,"Solar electricity generation (kWh)")
df_no_outliers = remove_outliers(df_no_outliers, "Electricity usage (kWh)")
```

```python
hourly_electricity_avg_clean = df_no_outliers.groupby("Hour").agg({
    
    'Solar electricity generation (kWh)': 'mean',
    'Electricity usage (kWh)': 'mean'
}).reset_index()
```

```python
#Create the plot 

plt.figure(figsize= (10,5))

plt.plot(hourly_electricity_avg_clean["Hour"],
        hourly_electricity_avg_clean["Solar electricity generation (kWh)"],
        label='Avg Solar Generation', marker='o')

plt.plot(hourly_electricity_avg_clean["Hour"],
        hourly_electricity_avg_clean["Electricity usage (kWh)"],
        label='Avg Solar Generation', marker='o')


plt.xlabel("Hour of Day")
plt.ylabel("Average Electricity (kWh) ")

plt.title("Average Solar Generation vs Electricity Usage by Hour (Outliers Removed)")
plt.legend()
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
#Now that data is cleaned; Calculate the electricity that was bought from provider to fill th deficit in 2020

df_no_outliers["Electricity bought (kWh) "] = df_no_outliers.apply(
    
    lambda row: max (0, row["Electricity usage (kWh)"] - row["Solar electricity generation (kWh)"]), axis =1

)# max 0 ensures that where electricity was not bought/no deficit it is replaced with 0 where the result is -ve

df_no_outliers.head(10)# The new column (Electricity bought (kWh))is the electricity deficit that needed to be bought to sustain the home

```

```python
# Calculate the excess solar generated 

df_no_outliers["Excess solar (kWh)"] = df_no_outliers.apply(

    lambda row: max(0, float(row["Solar electricity generation (kWh)"])- float(row["Electricity usage (kWh)"])), axis=1
)


```

```python
#Now lets assume a battery had been installed to store the excess solor generated electricity

battery_capacity = 12.5

battery_charge = 0 #as of 2020, 0000hrs, January

#convert the relevant columns to float

df_no_outliers["Electricity bought (kWh) "] =df_no_outliers["Electricity bought (kWh) "].astype(float)
df_no_outliers["Excess solar (kWh)"] =df_no_outliers["Excess solar (kWh)"].astype(float)

df_no_outliers['Battery charge (kWh)'] = 0.0


# calculate the battery charge level
for index, row in df_no_outliers.iterrows():
    battery_charge += row['Excess solar (kWh)'] - row["Electricity bought (kWh) "]
    battery_charge = max(0.0, min(battery_charge, battery_capacity))
    df_no_outliers.at[index, 'Battery charge (kWh)'] = battery_charge

    
```

```python
print(f"Average battery charge: {df_no_outliers['Battery charge (kWh)'].mean():.2f} kWh")
print(f"Maximum battery charge reached: {df_no_outliers['Battery charge (kWh)'].max():.2f} kWh")
```

```python
# Electricity bought with battery installed


df_no_outliers['Electricity bought with battery (kWh)'] = 0.0
battery_charge = 0.0

for index, row in df_no_outliers.iterrows():
    # Calculate required electricity
    required = float(row['Electricity usage (kWh)'])
    solar_gen = float(row['Solar electricity generation (kWh)'])

    # Use solar first
    needed_after_solar = max(0, required - solar_gen)

    # Use battery if available
    used_from_battery = min(battery_charge, needed_after_solar)
    needed_after_battery = needed_after_solar - used_from_battery

    # Update battery charge
    battery_charge -= used_from_battery #Reduces the battery charge by the amount used
    excess_solar = max(0, solar_gen - required)
    battery_charge = min(battery_capacity, battery_charge + excess_solar)
    
    """
    why use min() :If battery_charge is 5 kWh and excess_solar is 8 kWh, the new battery charge would be 13 kWh (5 + 8). 
    But since the battery's capacity is 12.5 kWh, the code will set battery_charge to 12.5 kWh.
    
    """

    

    # Electricity that needs to be bought
    df_no_outliers.at[index, 'Electricity bought with battery (kWh)'] = needed_after_battery


print(f"Electricity bought with battery installed:{df_no_outliers['Electricity bought with battery (kWh)'].head(10)}")


```

```python
total_bought_with_battery = df_no_outliers['Electricity bought with battery (kWh)'].sum()


print(f" The elecricity bought in 2020 with battery is: {total_bought_with_battery:.2f}(kWh)")
```
